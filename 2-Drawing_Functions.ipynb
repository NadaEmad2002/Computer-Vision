# Drawing Functions

- Draw a line
- Draw arrow segment
- Draw an ellipse
- Draw a circle
- Draw a rectangle
- Draw a text string
- Create Graphics

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

To create a black image

```python
img = np.zeros([512,512,3],np.uint8)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw a line
# Syntax:

cv2.line(image, start_point, end_point, color, thickness) 

```python
img=cv2.line(img,(0,0),(256,256),(255,0,0),5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw arrow segment
# Syntax: 

cv2.arrowedLine(image, start_point, end_point, color, thickness, line_type, shift, tipLength)

```python
img=cv2.arrowedLine(img,(0,255),(256,256),(0,255,0),5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw an ellipse
# Syntax: 

cv2.ellipse(image, centerCoordinates, axesLength, angle, startAngle, endAngle, color [, thickness[, lineType[, shift]]])

```python
img=cv2.ellipse(img,(250,460) ,(250, 60) ,0,0,360,(255, 0, 255) ,5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw a rectangle
# Syntax: 

   cv2.rectangle(image, start_point, end_point, color, thickness)

```python
img=cv2.rectangle(img,(322,0),(512,190),(0,0,255),-1)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw a circle
# Syntax: 

cv2.circle(image, center_coordinates, radius, color, thickness)

```python
img=cv2.circle(img,(256,256),150,(200,0,56),5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw a text string
# Syntax:

 cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])

```python
font=cv2.FONT_HERSHEY_SIMPLEX #تحديد نوع الخط 

img=cv2.putText(img,'opencv',(50,490),font,4,(0,100,155),5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Draw irregular shapes

```python
pts=np.array([[256,256],[270,200],[300,240],[400,300]],np.int32)# تحديد النقاط

img=cv2.polylines(img,[pts],True,(50,100,105),5)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Create Graphics

```python
img = np.zeros((600,900,3),dtype=np.uint8)

#background
cv2.rectangle(img,(0,0),(900,500),(255,255,85),-1)
cv2.rectangle(img,(0,500),(900,600),(75,180,70),-1)

#Sun
cv2.circle(img,(150,150),60,(0,255,255),-1)
cv2.circle(img,(150,150),75,(220,255,255),8)

# ***TREE 1 ***

#tree stem
cv2.line(img,(710,500),(710,420),(30,65,155),10)
#tree leafs
triangle=np.array([[640,460],[780,460],[710,200]],dtype=np.int32)
cv2.fillPoly(img,[triangle],(75,180,70))

# ***TREE 2 ***

#tree stem
cv2.line(img,(600,500),(600,420),(30,65,155),25)
#tree leafs
triangle=np.array([[500,440],[700,440],[600,75]],dtype=np.int32)
cv2.fillPoly(img,[triangle],(75,200,70))

#text
font = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText(img,'I Love Python',(120,490),font,1.5,(255,255,255),2)

cv2.imwrite('image.png',img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()
```

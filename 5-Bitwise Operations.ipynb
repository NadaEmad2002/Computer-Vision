# Bitwise Operations (AND,OR,NOT,XOR)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

create simple images

```python
img1=np.zeros((250,500,3),np.uint8)
img1=cv2.rectangle(img1,(200,50),(300,200),(255,255,255),-1)

img2=np.full((250,500,3),255,np.uint8)
img2=cv2.circle(img2,(156,150),100,(0,0,0),-1)

titles = ['Image 1','Image 2']
images = [img1,img2]

for i in range(2):
    plt.subplot(1,2,i+1),
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
bitAnd=cv2.bitwise_and(img1,img2)

plt.imshow(bitAnd)
plt.title('And Operator')
plt.show()
```

```python
bitOr=cv2.bitwise_or(img1,img2)

plt.imshow(bitOr)
plt.title('OR Operator')
plt.show()
```

```python
bitXor=cv2.bitwise_xor(img1,img2)

plt.imshow(bitXor)
plt.title('Xor Operator')
plt.show()
```

```python
bitNot1=cv2.bitwise_not(img1)
bitNot2=cv2.bitwise_not(img2)

titles = ['Image 1','Image 2','Not Operator Image 1','Not Operator Image 2']
images = [img1,img2,bitNot1,bitNot2]

for i in range(4):
    plt.subplot(2,2,i+1),
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

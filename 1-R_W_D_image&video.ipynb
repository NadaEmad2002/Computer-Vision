# Import Needed Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import datetime
```

# Read Image

# Syntax: 

cv2.imread( path , flag )

Parameters:

path: A string representing the path of the image to be read.

flag: It specifies the way in which image should be read.

     It is for the colors of the photo
      # 0  for grayscale
      # 1  for color

```python
# uploads a specific image
img = cv2.imread('Images/car.jpg',1)

# Print image pixels
print(img)
```

```python
# Displaying image using plt.imshow()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # We use cvtColor to convert the image from BGR to RGB
plt.title('Car')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img1 = img
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # We use cvtColor to convert the image from BGR to RGB
titles = ['BGR Image','RGB Image']
images = [img1,img2]

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

# Save Image
# Syntax: cv2.imwrite(filename, image)

Parameters:

filename: A string representing the file name. The filename must include image format like .jpg, .png, etc.

image: It is the image that is to be saved.

```python
cv2.imwrite('output.jpg', img)
```

# Capture Video from Camera

```python
cap=cv2.VideoCapture(0)# to get a video capture object for the camera.
fourcc=cv2.VideoWriter_fourcc(*'XVID')# to improve the vedio resolution

# to save vedio
out=cv2.VideoWriter('output.avi',fourcc,20.0,(640,480))

# Check if the camera is opened or not
print(cap.isOpened())
# Read the entire file until it is completed 
while(cap.isOpened()): 
    ret,frame=cap.read() # read every thing that camers show it 
    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out.write(frame)
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # to convert frame from RGB to Gray 
        cv2.imshow('frame',gray)

        if cv2.waitKey(1)&0xFF==ord('q'): # define 'q' as the exit button
            break
    else:
            break 
cap.release() # to close the camera
out.release()
cv2.destroyAllWindows() # Closes all the windows currently opened. 
```

```python
#Show Date and Time on Videos
cap=cv2.VideoCapture(0)
cap.set(3,3000)
cap.set(4,3000)
print(cap.get(3))
print(cap.get(4))
while(cap.isOpened()):
    ret,frame = cap.read()
    if ret==True:
        font=cv2.FONT_HERSHEY_COMPLEX
        text='Width: '+str(cap.get(3))+'Height: '+str(cap.get(4))
        datet=str(datetime.datetime.now())
        frame=cv2.putText(frame,text,(10,50),font,1,(0,255,255),2)
        frame=cv2.putText(frame,datet,(10,100),font,1,(0,255,255),2)
        cv2.imshow('Frame',frame)
        if cv2.waitKey(1)==ord('q'):
            break
    else:
            break
cap.release()
cv2.destroyAllWindows()        
```

```python
# Create a VideoCapture object and read from input file 
cap = cv2.VideoCapture('Images/walking.avi') 

# Check if camera opened successfully 
if (cap.isOpened()== False): 
    print("Error opening video file") 
  
# Read until video is completed 
while(cap.isOpened()): 
      
# Capture frame-by-frame 
    ret, frame = cap.read() 
    if ret == True: 
    # Display the resulting frame 
        cv2.imshow('Frame', frame) 
          
    # Press Q on keyboard to exit 
        if cv2.waitKey(25) & 0xFF == ord('q'): 
            break
  
# Break the loop 
    else: 
        break
  
# When everything done, release 
# the video capture object 
cap.release() 
  
# Closes all the frames 
cv2.destroyAllWindows()      
```

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img1=cv2.imread('Images/car.jpg')
img2=cv2.imread('Images/anime.jpg')
print(img1.size) 
print(img2.size) 
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2= cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

titles = ['Image 1','Image 2']
images = [img1,img2]

for i in range(2):
    plt.subplot(1,2,i+1),
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

# Merge two Image

```python
img1=cv2.resize(img1,(640,427))
img2=cv2.resize(img2,(640,427))

dst=cv2.add(img1,img2) #   add

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()
```

- we use addWeighted to control the merging ratio between the two images

```python
img1=cv2.resize(img1,(640,427))
img2=cv2.resize(img2,(640,427))

dst=cv2.addWeighted(img1,.25,img2,.75,0) #     addWeighted

plt.imshow((dst))
plt.xticks([]), plt.yticks([])
plt.show()
```
